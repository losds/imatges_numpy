- Exercici 5

Utilitza la Indexació per extreure els valors d'una columna i una fila de la matriu. I suma els seus valors.

```python
def sumaFilaCOlumna(matr, fila,columna):
    print("Matriz original",matr)
    filam=matr[fila,:]
    print(f'Fila seleccionada {fila} :',filam)
    columnam=matr[:,columna]
    print(f'Columna seleccionada {columna} :',columna)
    res=filam+columnam
    print('Resultado :',res)
```

```python
matr3x3=np.random.randint(0,20, size=(3, 3))
```

```python
matr3x3
```

```python
sumaFilaCOlumna(matr3x3,2,2)
```
